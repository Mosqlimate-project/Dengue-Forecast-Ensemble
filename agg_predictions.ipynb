```python
import numpy as np
import pandas as pd
from epiweeks import Week
```

```python
mandatory_states = ['AM', 'CE', 'PR', 'GO', 'MG']

dates_25 = pd.date_range(start= Week(2024, 41).startdate().strftime('%Y-%m-%d'),
              end= Week(2025, 40).startdate().strftime('%Y-%m-%d'),
              freq='W-SUN')
```

```python
df_forecast = pd.DataFrame()
for state in mandatory_states:

    df_for = pd.DataFrame()

    for model_id in [21,22,25,27,28,30,33]:

        df_ = pd.read_csv(f'./predictions/for_{model_id}_{state}_2025.csv.gz')
        df_.date = pd.to_datetime(df_.date)
        df_.set_index('date', inplace = True)
        df_ = df_[['pred', 'lower', 'upper']]
        df_.columns = df_.columns + f'_{model_id}'

        df_ = df_.loc[df_.index.isin(dates_25)]
        
        df_for = pd.concat([df_for, df_],axis =1)
        

    df_for['state'] = state

    df_forecast = pd.concat([df_forecast, df_for])

df_forecast.head()
```

```python
df_forecast.state.unique()
```

```python
df_forecast.to_csv('forecast_mandatory_2025.csv')
```
